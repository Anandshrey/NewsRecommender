## Perform Natural Language Processing

```python
# Remove Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

```python
stop_words = set(stopwords.words('english'))
for i in range(len(df_copy["TITLE"])):
    string = ""
    for word in df_copy["TITLE"][i].split():
        word = ("".join(e for e in word if e.isalnum()))
        word = word.lower()
        if not word in stop_words:
          string += word + " "
    if(i%1000==0):
      print(i)           # To track number of records processed
    df_copy.at[i,"TITLE"] = string.strip()
```

```python
# LEmmatize
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
nltk.download('punkt')
```

```python
for i in range(len(df_copy["TITLE"])):
    string = ""
    for w in word_tokenize(df_copy["TITLE"][i]):
        string += lemmatizer.lemmatize(w,pos = "v") + " "
    df_copy.at[i, "TITLE"] = string.strip()
    if(i%1000==0):
        print(i)
```

## Vectorize using TFID vectorizer

```python
tfidf_vectorizer = TfidfVectorizer(min_df = 0)
tfidf_features = tfidf_vectorizer.fit_transform(df_copy['TITLE'])
print(tfidf_features)
```

```python
print(tfidf_features.shape)
print(tfidf_features[0])
print(type(tfidf_features))
```

## Define Recommender

```python
# recommend based on simmilarity between a queried item and the news items using euclidean distance between vectors
def tfidf_based_model(row_index, num_similar_items):
    couple_dist = pairwise_distances(tfidf_features,tfidf_features[row_index])
    indices = np.argsort(couple_dist.ravel())[0:num_similar_items]
    df_new = pd.DataFrame({'TITLE':df['TITLE'][indices].values,
                'Euclidean Similarity with Queried Article': couple_dist[indices].ravel()})
    print("="*30,"Queried Article Title","="*30)
    print('TITLE : ',df['TITLE'][indices[0]])
    print("\n","="*29,"Recommended Articles","="*30)

    return df_new.iloc[1:,]
```
